```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

file_path = 'Data MoMo.xlsx'
xls = pd.ExcelFile(file_path)
df1 = pd.read_excel(xls, sheet_name=xls.sheet_names[0])
df2 = pd.read_excel(xls, sheet_name=xls.sheet_names[1])
df3 = pd.read_excel(xls, sheet_name=xls.sheet_names[2])

print("Sheet 1:")
print(df1)
print("\nSheet 2:")
print(df2)
print("\nSheet 3:")
print(df3)
```

```python
df1['Amount'] = df1['Amount'].str.replace(',', '').astype(float)
```

```python
def convert_date(date):
    try:
        return pd.to_datetime(date, format='%Y-%m-%d')
    except ValueError:
        return pd.to_datetime(date, format='%d/%m/%Y')

df1['Date'] = df1['Date'].apply(convert_date)
df1['Date'] = df1['Date'].dt.strftime('%Y-%m-%d')

print(df1)
```

```python
def convert_date(date):
    try:
        converted_date = pd.to_datetime(date, format='mixed', dayfirst=True)
        if converted_date.year > 2100:
            return None
        return converted_date
    except ValueError:
        return None

df3['First_tran_date'] = df3['First_tran_date'].apply(convert_date)

df3 = df3.dropna(subset=['First_tran_date'])

df3['First_tran_date'] = df3['First_tran_date'].dt.strftime('%Y-%m-%d')

print(df3)
```

##Combined with the 'User_Info' table, add columns: Age, Gender, Location, Type_user (New/Current) in 'Transactions' table

```python
warnings.filterwarnings("ignore")
df3.rename(columns={'User_id': 'user_id'}, inplace=True)
df1 = pd.merge(df1, df3[['user_id', 'Age', 'Gender', 'Location', 'First_tran_date']], on='user_id', how='left')

df1
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1['First_tran_date'] = pd.to_datetime(df1['First_tran_date'])
```

```python
df1['Type_user'] = df1.apply(lambda row: 'New' if (row['Date'].year == row['First_tran_date'].year) and (row['Date'].month == row['First_tran_date'].month) else 'Current', axis=1)
df1
```

```python
with pd.ExcelWriter('MoMo_Clean.xlsx') as writer:
    df1.to_excel(writer, sheet_name='Data Transactions', index=False)
    df2.to_excel(writer, sheet_name='Data Commission', index=False)
    df3.to_excel(writer, sheet_name='Data User_Info',  index=False)
```
